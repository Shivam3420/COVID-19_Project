```python
import pandas as pd 
import matplotlib.pyplot as plt

d = pd.read_csv('covid_19_india.csv')
df = pd.DataFrame(d)
print(df)

D = df[['State/UnionTerritory','Confirmed','Cured','Deaths']]
print(D)

dfp = pd.pivot_table(df,index="State/UnionTerritory").drop(columns = 'Sno').sort_values(by="Confirmed")
print(dfp)

dfp.plot(kind='bar')
plt.show()
```
